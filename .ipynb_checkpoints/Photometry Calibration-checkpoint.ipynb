# Photometry

```python
import glob
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm, SymLogNorm

%matplotlib inline
```

```python
# datadir = 'Photometry/'
# bias_subdir = 'Bias_Frames'
# bias_subdir = "/media/sf_Astro_461_shared/Project/Photometry/"
```

## Bias

```python
# bias_files = glob.glob("/media/sf_Astro_461_shared/Project/Photometry/Bias_Frames/"+'*.fits')
bias_files = glob.glob("Photometry/Bias_Frames/"+'*.fits')

bias_data = [fits.getdata(b) for b in bias_files]
master_bias = np.median(bias_data, axis=0)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
hist = ax1.hist(master_bias.flatten(), bins=1000, range=[2000,2500])
img = plt.imshow(master_bias, cmap='gray', vmin=2200, vmax=3000, origin='lower')#,ax=ax2)
plt.colorbar()
```

```python
master_bias_head = fits.getheader(bias_files[0])
fits.writeto('Photometry/'+'master_bias.fits', master_bias, header=master_bias_head, clobber=True)
```

## Flat

```python
flat_files_R = glob.glob("Photometry/Flat_Fields_R/"+'*.fits')

# Again, we need to remove the bias counts from all flat field exposures to 
# determine the contribution from just the flats
# flat_data = [fits.getdata(f)-master_bias for f in flat_files]
flat_data_R = [np.subtract(fits.getdata(f), master_bias) for f in flat_files_R]
master_flat_R = np.median(flat_data_R, axis=0)
master_flat_R = master_flat_R/np.median(master_flat_R)


flat_files_Ha13 = glob.glob("Photometry/Flat_Fields_Ha13/"+'*.fits')

# Again, we need to remove the bias counts from all flat field exposures to 
# determine the contribution from just the flats
# flat_data = [fits.getdata(f)-master_bias for f in flat_files]
flat_data_Ha13 = [np.subtract(fits.getdata(f), master_bias) for f in flat_files_Ha13]
master_flat_Ha13 = np.median(flat_data_Ha13, axis=0)
master_flat_Ha13 = master_flat_Ha13/np.median(master_flat_Ha13)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
hist = ax1.hist(master_flat_R.flatten(), bins=1000, range=[0.5,1.5])
img = plt.imshow(master_flat_R, cmap='gray',vmin=0.7, vmax=1.1, origin='lower')
plt.colorbar()
plt.show()


fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
hist = ax1.hist(master_flat_Ha13.flatten(), bins=1000, range=[0.5,1.5])
img = plt.imshow(master_flat_Ha13, cmap='gray',vmin=0.7, vmax=1.1, origin='lower')
plt.colorbar()
plt.show()
```

```python
master_flat_R_head = fits.getheader(flat_files_R[0])
fits.writeto("Photometry/"+'master_flat_R.fits', master_flat_R, 
             header=master_flat_R_head, clobber=True)

master_flat_Ha13_head = fits.getheader(flat_files_Ha13[0])
fits.writeto("Photometry/"+'master_flat_Ha13.fits', master_flat_Ha13, 
             header=master_flat_Ha13_head, clobber=True)
```
