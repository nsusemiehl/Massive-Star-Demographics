```r
library(tidyverse)
```

```r
can = read_csv("Candidates.csv")
can = can %>% mutate(rn = row_number()) %>% filter(rn > 1) %>% mutate(dec = as.numeric(`DEC \n(deg)`), ra = as.numeric(`RA \n(deg)`)) %>% select(TARGET, SP_CLASS, dec, ra)
can
```

```r
can %>% distinct(TARGET, ra, dec, SP_CLASS) %>% group_by(TARGET) %>% summarize(mean_ra = mean(ra), mean_dec = mean(dec)) %>% 
    filter((mean_ra > 150) & (mean_ra <  270) & (mean_dec > -50) & (mean_dec < 80))
```
